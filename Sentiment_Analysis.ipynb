```python
import pandas as pd

df = pd.read_csv("data/twitter_training.csv", header=None)
df.head()
df.columns = ['Id', 'Entity', 'Sentiment', 'Content']
df.head()
```

```python
df.columns
```

```python
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'#', '', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text

df['clean_text'] = df['Content'].apply(clean_text)

df[['Content', 'clean_text']].head()
```

```python
df.to_csv("data/processed_twitter.csv", index=False)
```
